# HTML 태그 제거하기

HTML 태그를 제거하기 위해서는 정규 표현식을 활용

정규 표현식이란 일정한 패턴을 문장 속에서 찾아내는 기능


```python
import re
```

```python
def remove_html(texts):

  pre_text = []
  for text in texts:
    text = re.sub(r"<[^>]+>", "", text).strip()
    if text:
      pre_text.append(text)
  return pre_text
```

```python
context = remove_html(context)
for i, sent in enumerate(context):
  print(i, sent)
```

# 정규화

```python
def remove_email(texts):
  pre_text = []
  for text in texts:
    text = re.sub(r"[a-zA-Z0-9+-_.]+@[a-zA-Z0-9-]+\.[a-zA-Z0-9-.]+", "", text).strip()
    if text:
      pre_text.append(text)
  return pre_text
```

```python
sents = remove_email(sents)
for i, sent in enumerate(sents):
  print(i, sent)
```

```python
def remove_hashtag(texts):
  pre_text = []
  for text in texts:
    text = re.sub(r"#\S+", "", text).strip()
    if text:
      pre_text.append(text)
  return pre_text
```

```python
sents = remove_email(sents)
for i, sent in enumerate(sents):
  print(i, sent)
```

```python
def remove_user_mention(texts):
  pre_text = []
  for text in texts:
    text = re.sub(r"@\w+", "", text).strip()
    if text:
      pre_text.append(text)
  return pre_text
```

```python
sents = remove_user_mention(sents)

for i, sent in enumerate(sents):
  print(i, sent)
```

```python
def remove_url(texts):
  pre_text = []
  for text in texts:
    text = re.sub(r"(http|https)?:\/\/\S+\b|www.(\w+\.)+\S*","",text).strip()
    text = re.sub(r"pic\.(\w\.)+\S*","",text).strip()
    if text:
      pre_text.append(text)
  return pre_text
```

```python
sents = remove_url(sents)
for i, sent in enumerate(sents):
  print(i, sent)
```

```python
def remove_bad_char(texts):
    bad_chars = {"\u200b": "", "“": '"', "”": '"', "\ufeff": ""}

    pre_text = []
    for text in texts:
        for bad_char in bad_chars:
            text = text.replace(bad_char, bad_chars[bad_char])
        # text = re.sub(r"[^\x00-\x7F]+", "", text)
        if text:
            pre_text.append(text)
    return pre_text
```

```python
sents = remove_bad_char(sents)
for i, sent in enumerate(sents):
    print(i, sent)
```

```python
def remove_press(texts):

    re_patterns = [
        r"\(.*?\+?\s*(?:경향|한겨레|미디어|데일리|한국|한경|위키트리)\s*\)",  # (언론사명)
        r"\[.*?\]",  # 대괄호 내용 제거 (예: [사진 제공])
        r"[-=]{2,}",  # == 또는 -- 제거
        r"\s+",  # 공백 정리
    ]

    preprocessed_text = []
    for text in texts:
        for re_pattern in re_patterns:
            text = re.sub(re_pattern, "", text).strip()
        if text:
            preprocessed_text.append(text)

    return preprocessed_text
```

```python
def remove_repeat_char(texts):
    bad_chars = {"\u200b": "", "“": '"', "”": '"', "\ufeff": ""}

    pre_text = []
    for text in texts:
        for bad_char in bad_chars:
            text = repeat_normalize(text, num_repeats=2).strip()
        if text:
            pre_text.append(text)
    return pre_text
```

```python
# 중복 문장 제거

from collections import OrderedDict

def remove_dup_sent(texts):
  texts = list(OrderedDict.fromkeys(sents))
  return texts
```

```python
def spacing_sent(texts):

    pre_text = []
    for text in texts:
       text=spacing(text)
       if text:
        pre_text.append(text)
    return pre_text
```

```python
!pip install git+https://github.com/ssut/py-hanspell.git

from hanspell import spell_checker

spelled_sent = spell_checker.check(sents)
```

```python
def spell_check_sent(texts):

    pre_text = []
    for text in texts:
      try:
        spelled_sent = spell_checker.check(text)
        checked_sent = spelled_sent.checked
        if checked_sent:
          pre_text.append(checked_sent)
      except:
        pre_text.append(text)
    return pre_text
```

```python
sents = spell_check_sent(sents)
for i, sent in enumerate(sents):
  print(i, sent)
```
